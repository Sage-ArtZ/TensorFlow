```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

### Creating a single layered neuron with shape 1

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

```python
### Compiling the neuron to compute the loss and updating the weights and baises hence forth with optimizer (back prop)
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
Xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
### training the model with the Input & Output with 500 iteration i.e epoch
```

```python
model.fit(Xs, ys, epochs=500)
```

```python
## predicting the 'y'
```

```python
model.predict([10])
```
